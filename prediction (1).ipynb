```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading Data

```python
netflix_dataset = pd.read_csv('netflix_titles.csv') 
netflix_dataset.head(10)
```

## Data Cleaning

```python
netflix_dataset[['director', 'cast', 'country']] = netflix_dataset[['director', 'cast', 'country']].replace(np.nan, "NA")
netflix_dataset = netflix_dataset.rename(columns= {"listed_in": "genre"})
netflix_dataset.head(10)
```

Remove any rows that doesn't have any genre listed
Director, cast and country is not necessary for our movie recommendation algorithm so we won't remove any rows where the data is missing. 



```python
netflix_dataset.dropna(subset = ["genre", "rating"], inplace = True)
```

Replace all the NaN values with NA



```python
netflix_dataset.fillna("NA", inplace=True)
netflix_dataset = netflix_dataset.replace(to_replace= "", value = "NA")
```

Removing any duplicate movies or TV Shows

```python
netflix_dataset = netflix_dataset.drop_duplicates(subset=['title', 'director', 'cast'], keep= 'first')
```

Extract the month added from the date

```python
data = netflix_dataset['date_added'].str.split(" ").str[0]
data = data.replace("", "NA")
netflix_dataset["month_added"] = data

netflix_dataset.head(10)
```

Dropping columns which are not needed for making the recommendation engines.

```python
netflix_dataset.drop(columns= ["show_id", "duration", "date_added", "release_year"], inplace=True)
netflix_dataset.head(10)
```

```python
import nltk
nltk.download('wordnet')
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# tokenizer = TreebankWordTokenizer()
stemmer = WordNetLemmatizer()
cv = CountVectorizer(stop_words = 'english', binary = False)
tokenizer = nltk.RegexpTokenizer(r"\w+")
```

In the following convert function, we are converting a string to numeric representation. We are tokening the string and lemmatizing it and then fitting it to our model. This converts our string to a vector using bag of words method.

```python
def convert(category):
  data = netflix_dataset[category]
  a = list()
  for i in data:
    a.append(tokenizer.tokenize(i))

  b = list()
  for item in a:
    b.append(" ".join(stemmer.lemmatize(token) for token in item))

  features = cv.fit_transform(b)
  features = list(features.toarray())
  new_netflix_dataset[category] = features
```

```python
final_df = new_netflix_dataset.copy()
final_df.drop("month_added", inplace = True, axis = 1)
final_df = final_df.set_index("title")
```

We have dropped the month added of movies as this will add noise to our data. 

```python
l = list()
for index, row in final_df.iterrows():
  val = list()
  val.extend([row['type']])
  val.extend(row['director'])
  val.extend(12*row['cast'])
  val.extend(5*row['country'])
  val.extend([row['rating']])
  val.extend(10*row['genre'])
  val.extend(10*row['description'])
  l.append(val)

```

We are assigning appropriate weightages to our features

```python
title = final_df.index
print(title)
data = {"title":  title, "vectorization": l}
vec_df = pd.DataFrame(data)
vec_df.set_index("title", inplace = True)
```

```python
vec_df
```

We have convert each row our dataframe into a single vector, so we can use it in our machine learning model. We are using the K nearest neighbour (KNN) method. To calculate the similarity between two vectors of our dataframe, we are using cosine similarity which calculates similarity by measuring the cosine of angle between two vectors. We can then return the movies and TV shows with hgihet similarity. Given the nature of our data, the KNN method seemed the most suitable choice. KNN is a method of unsupervised learning and it return the most similar movies/Tv shows.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def top_10(name):
    temp_df = vec_df.copy()
    vector = list(vec_df.loc[name])
    list_vect = []
    
    for index, row in vec_df.iterrows():
        vector1 = list(vec_df.loc[index])
        sim = cosine_similarity(vector, vector1)
        list_vect.append(sim[0][0])
    
    temp_df["cosine_similarity"] = list_vect
    top11 = temp_df.nlargest(11, ['cosine_similarity'])
    R_list = list(top11.index)
    return R_list
```

```python
def recommend(name):
    ret = top_10(name)
    netflix_dataset_new_index = netflix_dataset.copy()
    netflix_dataset_new_index = netflix_dataset_new_index.set_index("title")
    recommended = netflix_dataset_new_index.loc[[ret[1],ret[2],ret[3],ret[4],ret[5],ret[6],ret[7],ret[8],ret[9],ret[10]]]
    return recommended
```

We just have to pass a movie or TV show name to the function recommend which will then return us recommentations about movies and the data about those movies to help user chose what they want to watch. The suggestions for movie Zombieland can be seen below. Each movie/TV show name will return their correspondong similar recommendations.

```python
recommended_movies_shows = recommend("Zombieland")
recommended_movies_shows
```
